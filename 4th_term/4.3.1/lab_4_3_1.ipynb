```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
num_lines = np.array([2, 3, 4, 5, 6])
# num_lines = np.array([2, 3, 4, 5, 6])
micro_coords =  (53.9 - np.array([51.7, 52.7, 53.0, 53.2, 53.3])) * 10
# micro_coords =  (np.array([40.5, 40.2, 40, 39.9, 39.8]) - 39.3) * 10
num_zones = 1 / num_lines
```

```python
def get_line (x, y, delta=0.1):
    model = LinearRegression().fit(x.reshape(-1, 1), y)
    
    k = model.coef_[0]
    b = model.intercept_

    X = np.arange (x.min(), x.max() + delta, delta)
    Y = model.predict(X.reshape(-1, 1))

    return (X, Y), k, b
```

```python
(X, Y), k, b = get_line(num_zones, micro_coords, 0.0667)

# X = np.arange (num_zones.min(), num_zones.max() + 0.05, 0.05)
# Y = k*X + b

b_sign = '+' if b >= 0 else ''
plt.scatter (num_zones, micro_coords)
plt.plot(X, Y, label=f"m = {k:.2f}*z{b_sign}{b:.2f}")
plt.title (f"Зависимость m(z)")
plt.xlabel("1 \ m")
plt.ylabel("z, mm.")


plt.grid()
plt.legend()

plt.savefig (f"plot_fren.png")
plt.show()
```
