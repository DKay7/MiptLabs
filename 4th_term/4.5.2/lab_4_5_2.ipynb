```python
df = pd.read_csv("452.csv")
df.head()
```

```python
data = df.to_numpy()
```

```python
print('''
\\begin{table}[h]
\\centering
\\begin{tabular}{|c|c|c|c|c|c|c|c|c|}
\\hline ''')
print("№ & угол $\\beta$, град. & $cos(\\beta)$ & $\sigma cos(\\beta)$ & $\\gamma$ & $\\delta$ & $\\gamma_1$ & $\\gamma_3$ & $\\sigma \\gamma_3$ \\\ \\hline")

for idx, data_line  in enumerate(data):
    beta, cos_b, d_cos_b, g, delta, g1, g3, d_g3 = data_line
    print(f"{idx + 1} & {beta} & {cos_b:.3f} & {d_cos_b:.3f} & {g:.3f} & {delta:.3f} & {g1:.3f} & {g3:.3f}  & {d_g3:.3f} \\\ \\hline")

print('''
\end{tabular}
\caption{Расчитанные коэффициенты $\\gamma$, $\\gamma_1$, $\\delta$, $\\gamma_3$ и $cos(\\beta)$.}
\label{tabl:gammas}
\end{table}
''')
```

```python
cosinuses = []
d_cosinuses = []
gammas3 = []
d_gammas3 = []

for idx, data_line  in enumerate(data):
    beta, cos_b, d_cos_b, g, delta, g1, g3, d_g3 = data_line
    cosinuses.append(cos_b)
    d_cosinuses.append(d_cos_b)
    gammas3.append(g3)
    d_gammas3.append(d_g3)

cosinuses   = np.array(cosinuses)
d_cosinuses = np.array(d_cosinuses)
gammas3     = np.array(gammas3)
d_gammas3   = np.array(d_gammas3)
```

```python
plt.figure()
plt.errorbar(cosinuses, gammas3, xerr=d_cosinuses, yerr=d_gammas3, fmt='o', color="black", ecolor='gray')
plt.xlabel("cos(β)")
plt.ylabel("γ₃")
plt.grid()
# plt.show()

plt.savefig("gamma_from_cos_beta.png")
```

```python
print('''
\\begin{table}[h]
\\centering
\\begin{tabular}{|c|c|c|c|c|c|c|c|c|c|c|}
\\hline ''')
print("№ &  $x$, см. & $h_1$, дел.  & $h_2$, дел.  & $h_3$, дел. & $h_4$, дел. & $\\gamma$ & $\\delta$ & $\\gamma_1$ & $\\gamma_2$ & $\\sigma \\gamma_2$ \\\ \\hline")

for idx, data_line  in enumerate(data):
    x, h1, h2, h3, h4, g, delta, g1, g2, d_g2 = data_line
    print(f"{idx + 1} & {x} & {h1:} & {h2:} & {h3} & {h4} & {g:.3f} & {delta:.3f} & {g1:.3f} & {g2:.3f}  & {d_g2:.3f} \\\ \\hline")

print('''
\end{tabular}
\caption{Расчитанные коэффициенты $\\gamma$, $\\gamma_1$, $\\delta$, $\\gamma_2$.}
\label{tabl:x_to_gammas}
\end{table}
''')
```

```python
xs          = []
d_xs        = [1] * len(data)
gammas2     = []
d_gammas2   = []

for idx, data_line  in enumerate(data):
    x, h1, h2, h3, h4, g, delta, g1, g2, d_g2 = data_line
    xs.append(x)
    gammas2.append(g2)
    d_gammas2.append(d_g2)

xs          = np.array(xs       )
d_xs        = np.array(d_xs     )
gammas2     = np.array(gammas2  )
d_gammas2   = np.array(d_gammas2)
```

```python
plt.figure()
plt.errorbar(xs, gammas2, xerr=d_xs, yerr=d_gammas2, fmt='.', color="black", ecolor='gray')
plt.xlabel("x, см.")
plt.ylabel("γ₂")
plt.grid()
# plt.show()

plt.savefig("gamma_from_x.png")
```
