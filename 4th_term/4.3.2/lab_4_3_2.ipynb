```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def get_line (x, y, delta=0.1):
    model = LinearRegression().fit(x.reshape(-1, 1), y)
    
    k = model.coef_[0]
    b = model.intercept_

    X = np.arange (x.min(), x.max() + delta, delta)
    Y = model.predict(X.reshape(-1, 1))

    return (X, Y), k, b
```

```python
dataX1 = np.array([1, 2, 3, 4, 5, 6, 7])
dataY1 = np.array([57, 88, 120, 150, 181, 214, 244]) * 4

dataX2 = np.array([1, 2, 3, 4, 5, 6, 7])
dataY2 = np.array([60, 88, 121, 153, 190, 226, 258]) * 4


dataX3 = np.array([1, 2, 3, 4, 5])
dataY3 = np.array([85, 118, 157, 195, 233]) * 4
```

```python
def plot_data(dataX, dataY, savename):
    (X, Y), k, b = get_line(dataX, dataY)
    
    b_sign = '+' if b >= 0 else ''
    
    plt.plot(X, Y, color="grey", label=f"x(m) = {k:.2f}*m{b_sign}{b:.2f}")
    plt.scatter(dataX, dataY, color="black", label='Исходные данные')
    plt.grid()
    plt.xlabel("m")
    plt.ylabel("x, мкм")
    plt.legend()
    plt.savefig(savename)
    plt.show()
```

```python
plot_data(dataX1, dataY1, "graph1.png")
plot_data(dataX2, dataY2, "graph2.png")
plot_data(dataX3, dataY3, "graph3.png")
```
