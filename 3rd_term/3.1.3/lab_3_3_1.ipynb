```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
n = np.array([12, 11, 10, 9, 8, 7, 6, 5, 4])
T = np.array([3.4, 3.1, 2.8, 2.7, 2.25, 1.95, 1.6, 1.45, 1.15])
```

```python
def get_line (x, y, delta=0.1):
    model = LinearRegression().fit(x.reshape(-1, 1), y)
    
    k = model.coef_[0]
    b = model.intercept_

    X = np.arange (x.min(), x.max() + delta, 0.1)
    Y = model.predict(X.reshape(-1, 1))

    return (X, Y), k, b
```

```python
(X, Y), k, b = get_line(n, T)

plt.figure(figsize=(12, 8))
plt.plot(X, Y, label="Аппроксимированное T(n)")
plt.scatter(n, T, label="T(n)")

plt.xlabel("n")
plt.ylabel("T, с.")
plt.title("Зависимость периода колебаний от количества магнитов")

plt.legend()
plt.grid()

plt.savefig("plotBh.png")
plt.show()
```

```python
n2 = np.array([12, 10, 8, 6, 4])
M = np.array([470, 470.4, 405.68, 340.04, 241.08])
```

```python
(X, Y), k, b = get_line(n2, M)

plt.figure(figsize=(12, 8))
plt.plot(X, Y, label="Аппроксимированное M(n)")
plt.scatter(n2, M, label="M(n)")

plt.xlabel("n")
plt.ylabel("M, дин * см.")
plt.title("Зависимость момента механических сил от количества магнитов")

plt.legend()
plt.grid()
plt.savefig("plotBv.png")
plt.show()
```
