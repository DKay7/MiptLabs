```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
tau = np.array([50, 75, 100, 125, 150, 175, 200])
d_nu = np.array([19.6, 13.4, 9.8, 8, 6.5, 5.5, 4.5])
inverse_tau = 1 / tau
```

```python
def get_line (x, y):
    delta= (x.max() - x.min()) / 100

    model = LinearRegression().fit(x.reshape(-1, 1), y)
    
    k = model.coef_[0]
    b = model.intercept_

    X = np.arange (x.min(), x.max() + delta, delta)
    Y = model.predict(X.reshape(-1, 1))

    return (X, Y), k, b
```

```python
(X, Y), k, b = get_line(inverse_tau, d_nu)

plt.figure(figsize=(12, 8))
plt.plot(X, Y, label="Аппроксимированное Δν(1 / τ)")
plt.scatter(inverse_tau, d_nu, label="Δν(1 / τ)")

plt.xlabel("1 / τ, мкс ^ (-1)")
plt.ylabel("Δν, Гц")
plt.title("Зависимость Δν(1 / τ)")

plt.legend()
plt.grid()

plt.savefig("plot_A1.png")
plt.show()
```
